## 1. Variabel Temperature

### Visualisasi Data dan Analisis Distribusi

```python
def histogram(data, column, bins=30):
    plt.figure(figsize=(10, 6)) 
    plt.subplot() 
    sns.histplot(data[column], bins=bins, kde=True) 
    plt.title(f'Histogram {column}')
    plt.xlabel(column)
    plt.ylabel('Frekuensi')
    plt.show()  

histogram(df, 'Temperature')
```

Berdasarkan visualisasi diatas, diketahui bahwa variabel Temperature tergolong dalam Distribusi Bimodal karena memiliki dua puncak, yaitu antara suhu -20 hingga 0 derajat serta suhu 20 hingga 40 derajat. Hal ini dapat disebabkan oleh pengaruh variabel lain dalam dataset, seperti lokasi pengamatan serta musim pada saat pengamatan dilakukan. 

### Ukuran Pemusatan Data, Penyebaran Data dan Analisis Skewness

```python
def calculate_statistics(data, column_name):
    # Ukuran Pemusatan
    mean = data[column_name].mean()
    median = data[column_name].median()
    modus = data[column_name].mode()[0] 
    print('Ukuran Pemusatan variabel', column_name, ':')
    print('Mean:', mean)
    print('Median:', median)
    print('Modus:', modus)

    # Ukuran Penyebaran
    varians = data[column_name].var()
    std_dev = data[column_name].std()
    q1 = data[column_name].quantile(0.25)
    q2 = data[column_name].quantile(0.5)  
    q3 = data[column_name].quantile(0.75)
    range_data = data[column_name].max() - data[column_name].min()
    print('\nUkuran Penyebaran variabel', column_name, ':')
    print('Varians:', varians)
    print('Standar Deviasi:', std_dev)
    print('Q1:', q1)
    print('Q2:', q2)
    print('Q3:', q3)
    print('Range:', range_data)
    
    # Menghitung Skewness
    skewness = data[column_name].skew()
    print('\nSkewness dari variabel', column_name, ':', skewness)

calculate_statistics(df, 'Temperature')
```

1. Ukuran Pemusatan Data

Dari perhitungan yang telah dilakukan, diketahui bahwa rata-rata variabel Temperature dalam dataset berada di kisaran 19.13 derajat Celsius. Nilai tengah atau mediannya menunjukkan bahwa setengah dari pengukuran suhu berada di bawah 21 derajat dan setengahnya lagi di atas 21 derajat Celsius. Sementara itu, modus mengindikasikan bahwa suhu 32 derajat adalah nilai yang paling sering muncul dalam dataset. 

2. Ukuran Penyebaran Data

Varians sebesar 302.28 dan standar deviasi 17.39 menandakan bahwa variabel Temperature dalam dataset memiliki penyebaran nilai yang cukup besar, serta masing-masing nilai menyimpangan sekitar 17.39 derajat dari rata-rata keseluruhan. Kuartil pertama (Q1) yang berada di nilai 4 derajat Celsius menunjukkan bahwa 25% dari pengukuran suhu berada di bawah 4 derajat celcius. Kuartil kedua (Q2) yang berada di nilai 21 derajat menunjukkan titik tengah distribusi suhu, sementara kuartil ketiga (Q3) di 31 derajat menunjukkan bahwa 75% dari pengukuran suhu berada di bawah angka ini. Rentang yang mencapai 134.0 derajat menunjukkan perbedaan yang sangat besar antara suhu terendah dan tertinggi yang ada dalam dataset.

3. Analisis Skewness

Skewness sebesar 0.22 menunjukkan bahwa variabel temperature memiliki skewness positif dan distribusinya cenderung miring ke arah kanan. ## 1. Variabel Cloud Cover

### Visualisasi Data dengan Diagram Batang

```python
def bar_chart(data, column):
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x=column)
    plt.title(f'Diagram Batang untuk {column}')
    plt.xlabel(column)
    plt.ylabel('Frekuensi')
    plt.xticks(rotation=45) 
    plt.show()

bar_chart(df, 'Cloud Cover')
```

### Frekuensi dan Proporsi Data

```python
def categorical_variable(data, column):
    frekuensi = data[column].value_counts()
    proporsi = data[column].value_counts(normalize=True)

    print('Frekuensi setiap kategori', column, ':')
    print(frekuensi)
    print('\nProporsi setiap kategori', column, ':')
    print(proporsi)

categorical_variable(df, 'Cloud Cover')
```

1. Frekuensi Setiap Kategori

Dari perhitungan tersebut, diketahui bahwa kategori overcast memiliki jumlah pengamatan yang paling tinggi, yaitu 6090. Angka ini kemudian diikuti oleh kategori partly cloudy dengan pengamatan sebanyak 4560, lalu clear sebanyak 2139, serta cloudy sebanyak 411. Dengan begitu, dapat disimpulkan bahwa kondisi overcast menjadi kategori cloud cover yang paling sering tercatat dalam dataset, sedangkan kondisi cloudy memiliki frekuensi yang paling sedikit.

2. Proporsi Setiap Kategori

Proporsi menunjukkan perbandingan setiap kategori terhadap total keseluruhan. Dari data ini, overcast memiliki proporsi terbesar, yaitu 46.14% dari total pengamatan. Persentase tersebut diikuti oleh kategori partly cloudy dengan 34.54%, lalu clear dengan 16.20%, serta cloudy dengan 3.11%.
